# Dataset Overview

This dataset contains detailed information on various criminal incidents reported by law enforcement agencies across different states in the U.S. Each row in the dataset corresponds to a specific crime incident, providing insights into the location, time, type of crime, as well as details about the victim(s) and perpetrator(s). The key columns and their descriptions are as follows:

### Key Columns:

1. **Record ID**: A unique identifier assigned to each crime incident.
2. **Agency Code**: The code of the law enforcement agency that reported the crime.
3. **Agency Name**: The name of the law enforcement agency responsible for reporting the crime (e.g., "Anchorage Municipal Police").
4. **Agency Type**: Describes the type of agency (e.g., "Municipal Police") to show the category of the reporting organization.
5. **City** & **State**: These columns provide the geographic location of the crime.
6. **Year** & **Month**: The year and month when the incident occurred.
7. **Incident**: Identifies the crime incident within the given year and location.
8. **Crime Type**: The type of crime committed (e.g., "Murder or Manslaughter"), providing a clear categorization of the incident.
9. **Victim Ethnicity**: Describes the ethnicity of the victim involved in the crime.
10. **Perpetrator Sex**, **Perpetrator Age**, **Perpetrator Race**, **Perpetrator Ethnicity**: Provide details about the perpetrator's demographics, including their sex, age, race, and ethnicity.
11. **Relationship**: The relationship between the victim and the perpetrator, for example, "Acquaintance" or "Stranger".
12. **Weapon**: The type of weapon used during the crime (e.g., "Blunt Object", "Strangulation").
13. **Victim Count**: The number of victims involved in the incident.
14. **Perpetrator Count**: The number of perpetrators involved in the incident.
15. **Record Source**: The source of the data, typically indicating where the crime report originated (e.g., "FBI").

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import folium
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import folium

df = pd.read_csv('/content/homicide_data_cleaned.csv')

state_homicides = df.groupby('State')['Victim Count'].sum().reset_index()


state_homicides['Victim Count'].fillna(0, inplace=True)


geojson_url = 'https://raw.githubusercontent.com/PublicaMundi/MappingAPI/master/data/geojson/us-states.json'
gdf = gpd.read_file(geojson_url)


merged = gdf.merge(state_homicides, left_on="name", right_on="State")


merged = merged[~merged['name'].isin(['Alaska', 'Hawaii'])]


fig, ax = plt.subplots(1, 1, figsize=(10, 8))


merged.boundary.plot(ax=ax, edgecolor='black')

# Plot the homicide data, coloring based on the 'Victim Count' column
merged.plot(column='Victim Count', cmap='Reds', linewidth=0.8, ax=ax, edgecolor='gray', legend=True)
plt.title('Homicide Counts by State')
plt.show()

# Step 6: Create an interactive map using Folium
m = folium.Map(location=[37.8, -96], zoom_start=4)

# Add the choropleth layer with homicide counts
folium.Choropleth(
    geo_data=geojson_url,
    name='choropleth',
    data=state_homicides,
    columns=['State', 'Victim Count'],
    key_on='feature.properties.name',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Homicide Counts by State'
).add_to(m)

# Add state boundaries to the map
folium.LayerControl().add_to(m)

# Save the interactive map to an HTML file
m.save('homicide_by_state_map.html')

# To view the map, open the generated HTML file in a browser
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc, classification_report
import matplotlib.pyplot as plt
import seaborn as sns


X = data[['Victim Age', 'Perpetrator Age', 'Weapon']]  # select relevant features
y = data['Crime Solved'].apply(lambda x: 1 if x == 'Yes' else 0)  # convert to binary

# Fill missing values in 'Weapon' with 'Unknown'
X['Weapon'].fillna('Unknown', inplace=True)

# Define preprocessing for categorical and numeric data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['Victim Age', 'Perpetrator Age']),  # scale numeric features
        ('cat', OneHotEncoder(handle_unknown='ignore'), ['Weapon'])  # one-hot encode categorical features
    ])

# Define the Random Forest model pipeline
rf_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(n_estimators=100, random_state=42))
])

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the Random Forest pipeline
rf_pipeline.fit(X_train, y_train)

# Make predictions with Random Forest
y_pred_rf = rf_pipeline.predict(X_test)

# Evaluate Random Forest accuracy
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Random Forest Accuracy: {accuracy_rf * 100:.2f}%")

# Confusion Matrix
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(conf_matrix_rf, annot=True, fmt="d", cmap="Blues")
plt.title('Random Forest Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Classification report
print("Random Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))

# ROC Curve for Random Forest
y_pred_prob_rf = rf_pipeline.predict_proba(X_test)[:, 1]  # Probability of the positive class
fpr_rf, tpr_rf, thresholds_rf = roc_curve(y_test, y_pred_prob_rf)
roc_auc_rf = auc(fpr_rf, tpr_rf)

# Plot ROC Curve for Random Forest
plt.figure()
plt.plot(fpr_rf, tpr_rf, color='blue', lw=2, label=f'Random Forest (AUC = {roc_auc_rf:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your data is in a DataFrame called `data`
# Feature selection
X = data[['Victim Age', 'Perpetrator Age', 'Weapon']]  # select relevant features
y = data['Crime Solved'].apply(lambda x: 1 if x == 'Yes' else 0)  # convert to binary

# Fill missing values in 'Weapon' with 'Unknown'
X['Weapon'].fillna('Unknown', inplace=True)

# Define preprocessing for categorical and numeric data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['Victim Age', 'Perpetrator Age']),  # scale numeric features
        ('cat', OneHotEncoder(handle_unknown='ignore'), ['Weapon'])  # one-hot encode categorical features
    ])

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the Random Forest model pipeline
rf_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(n_estimators=100, random_state=42))
])

# Define the Logistic Regression model pipeline
logreg_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(random_state=42, max_iter=1000))
])

# Perform cross-validation on Random Forest model
cv_scores_rf = cross_val_score(rf_pipeline, X_train, y_train, cv=5)
print(f"Random Forest Cross-Validation Accuracy: {cv_scores_rf.mean() * 100:.2f}%")

# Perform cross-validation on Logistic Regression model
cv_scores_logreg = cross_val_score(logreg_pipeline, X_train, y_train, cv=5)
print(f"Logistic Regression Cross-Validation Accuracy: {cv_scores_logreg.mean() * 100:.2f}%")

# Now let's train and evaluate the Random Forest model on the test set
rf_pipeline.fit(X_train, y_train)
y_pred_rf = rf_pipeline.predict(X_test)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Random Forest Test Accuracy: {accuracy_rf * 100:.2f}%")

# Confusion Matrix for Random Forest
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(conf_matrix_rf, annot=True, fmt="d", cmap="Blues")
plt.title('Random Forest Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# ROC Curve for Random Forest
y_pred_prob_rf = rf_pipeline.predict_proba(X_test)[:, 1]  # Probability of the positive class
fpr_rf, tpr_rf, thresholds_rf = roc_curve(y_test, y_pred_prob_rf)
roc_auc_rf = auc(fpr_rf, tpr_rf)

# Plot ROC Curve for Random Forest
plt.figure()
plt.plot(fpr_rf, tpr_rf, color='blue', lw=2, label=f'Random Forest (AUC = {roc_auc_rf:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()

# Now let's train and evaluate the Logistic Regression model on the test set
logreg_pipeline.fit(X_train, y_train)
y_pred_logreg = logreg_pipeline.predict(X_test)
accuracy_logreg = accuracy_score(y_test, y_pred_logreg)
print(f"Logistic Regression Test Accuracy: {accuracy_logreg * 100:.2f}%")

# Confusion Matrix for Logistic Regression
conf_matrix_logreg = confusion_matrix(y_test, y_pred_logreg)
sns.heatmap(conf_matrix_logreg, annot=True, fmt="d", cmap="Blues")
plt.title('Logistic Regression Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# ROC Curve for Logistic Regression
y_pred_prob_logreg = logreg_pipeline.predict_proba(X_test)[:, 1]
fpr_logreg, tpr_logreg, thresholds_logreg = roc_curve(y_test, y_pred_prob_logreg)
roc_auc_logreg = auc(fpr_logreg, tpr_logreg)

# Plot ROC Curve for Logistic Regression
plt.figure()
plt.plot(fpr_logreg, tpr_logreg, color='green', lw=2, label=f'Logistic Regression (AUC = {roc_auc_logreg:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```
